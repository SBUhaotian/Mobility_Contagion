```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import optimize
```

```python
STATE_S = 100 # Suseptible
STATE_E = 101 # Exposed
STATE_I = 102 # Infected
STATE_C = 103 # Cared for / Isolated
STATE_R = 104 # Recovered
state_str = {STATE_S:'STATE_S', STATE_E:'STATE_E', STATE_I:'STATE_I', STATE_C:'STATE_C', STATE_R:'STATE_R'}

class Configs:
    def __init__(self, infect_prob, recover_time):
        self._infect_prob = infect_prob
        self._recover_time = recover_time
    
    def print(self, *argv):
        if (self._DBG == True):
            print ("["+inspect.stack()[1][3] + "] ", *argv)

    
    def __str__(self):
        return "_num_seeds:"+str(self._num_seeds)+"\n" +\
               "_reset_interval:" + str(self._reset_interval)+"\n"+\
               "_infect_prob:"+str(self._infect_prob)+"\n"+\
               "_asymp_prob:"+str(self._asymp_prob)+"\n"

class User:
    next_state_table = {STATE_S:STATE_E, STATE_E:STATE_I, STATE_I:[STATE_C, STATE_R],\
                        STATE_C:STATE_R, STATE_R:None}
    
    def __init__(self, config):
        self.state = STATE_S
        self.config = config
        self.group = None # This is only related when grouping
        self.recover_time = 0 # when she is recovered
        self.infected_from_me = 0
    
    def _validate_state(self):
        if (self.state not in [STATE_S, STATE_E, STATE_I, STATE_C, STATE_R]):
            print ("Unknown state:", self.state)
            raise Exception

    def sample_time(self, mean, std, left_bound):
        return mean
        sample = np.random.normal(loc = mean, scale=std)
        if (sample < left_bound):
            sample = left_bound
        return sample
    
    def _implicit_state_change(self, curtime):
        if (self.state == STATE_E and self.infectious_time > 0  and curtime >= self.infectious_time):
            self.state = STATE_I
        if (self.state == STATE_I and self.recover_time > 0 and curtime >= self.recover_time):
            self.state = STATE_R

    def expose(self, curtime, is_seed = False):
        if (self.state != STATE_S):
            raise Exception("Infecting with state:"+str(self.state))

        self.expose_time = curtime
        self.state = STATE_E
        #if (is_seed == True):
        #    self.infectious_time = curtime
        #else:
        self.infectious_time = curtime+1
        self.recover_time = self.infectious_time + self.config._recover_time
        self._implicit_state_change(curtime)


    def get_state(self, curtime):
        self._implicit_state_change(curtime)
        return self.state


    def __str__(self):
        # STATE_E-> infected STATE_I-> infectious
        return state_str[self.state]+ " is_symp:"+str(self.is_symp)+" expose_time:"+str(self.expose_time)+\
               " infectious_time:"+str(self.infectious_time)+" incare_time:"+str(self.incare_time)+\
               " recover_time:"+str(self.recover_time)+ "infected_from_me:"+str(self.infected_from_me)+"\n"
        

    def __repr__(self):
        return self.__str__()
```

```python
def simulate_by_contacts(N, beta, recover_time, nseed, infect_prob):
    user_dict = {}
    userids = np.arange(N, dtype='int')
    config = Configs(infect_prob=infect_prob, recover_time = recover_time)
    seeds = [0] #np.random.choice(np.arange(N, dtype='int'), size=nseed)
    for userid in userids:
        user_dict[userid] = User(config)
    for sid in seeds:
        user_dict[sid].expose(0, is_seed=True)
    ndays = 20
    new_infects = np.zeros(ndays)
    actives = []
    for d in range(ndays):
        for userid in userids:
            user_state = user_dict[userid].get_state(d)
            if (user_state == STATE_R):
                continue
            contacts = np.random.choice(userids, size=beta, replace=False)
            for contactid in contacts:
                contact_state = user_dict[contactid].get_state(d)
                sid = -1; iid = -1
                if (contact_state == STATE_S and user_state == STATE_I):
                    sid = contactid; iid = userid
                if (user_state == STATE_S and user_state == STATE_I):
                    sid = userid; iid = contactid
                if (sid != -1 and iid != -1):
                    if (np.random.uniform() <= infect_prob):
                        user_dict[sid].expose(d)
                        user_dict[iid].infected_from_me += 1
                        new_infects[d] += 1
        active = 0
        for userid in userids:
            if (user_dict[userid].get_state(d) == STATE_I):
                active+=1
        actives.append(active)
        #print (d, "active:", active)
    return new_infects, actives

_, axarr = plt.subplots(ncols=3, figsize=(14,3))
for beta in [1]:
    N=100; recover_time = 1; nseed = 1
    infect_prob = 1.0
    new_infects, actives = simulate_by_contacts(N, beta, recover_time, nseed, infect_prob)
    #print(new_infects)

    axarr[0].plot(np.cumsum(new_infects), label='$\\beta$='+str(beta));
    axarr[1].plot(new_infects)
    axarr[2].plot(actives)
axarr[0].legend()
plt.show()
```
